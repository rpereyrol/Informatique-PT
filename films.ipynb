Écrire une requête SQL qui renvoie le titre des westerns sortis depuis 1980.

```sql
SELECT *
FROM films
WHERE genre = "Western"
    AND annee >= 1980
```

Écrire une requête SQL qui renvoie les titres de films dont le résumé parle d'ingénieur ou de tennis.

```sql
SELECT *
FROM films
WHERE resume like "%tennis%"
    or resume like "%ingénieur%" 
```

Écrire une requête SQL qui renvoie les noms et prénoms des dix plus vieux artistes référencés.

```sql
SELECT *
FROM artistes
ORDER BY anneeNaiss
LIMIT 10
```

Écrire une requête SQL qui renvoie les noms de rôles joués par plusieurs acteurs différents.

```sql
SELECT nomRole,
    count(DISTINCT idActeur) as nb
FROM roles
GROUP BY nomRole
HAVING nb > 1
ORDER BY nb DESC
```

Écrire une requête SQL qui renvoie le nom et le prénom des réalisateurs.

```sql
SELECT DISTINCT a.nom,
    a.prenom
FROM artistes a
    JOIN films f ON a.idArtiste = f.idRealisateur
```

Écrire une requête SQL qui renvoie le nom et le prénom des acteurs qui ont joué le rôle d'un Skywalker.

```sql
SELECT DISTINCT a.nom,
    a.prenom
FROM artistes a
    JOIN roles r ON a.idArtiste = r.idActeur
WHERE r.nomRole like "%Skywalker%"
```

Écrire une requête SQL qui renvoie le nom et le prénom des réalisateurs qui ont réalisé leur premier film (référencé) avant l'âge de $30$ ans. Renvoyer aussi leur âge au moment de ce premier film.

```sql
SELECT a.nom,
    a.prenom,
    min(f.annee - a.anneeNaiss) as age
FROM artistes a
    JOIN films f ON a.idArtiste = f.idRealisateur
GROUP BY a.idArtiste
HAVING age <= 30
ORDER BY age
```

Écrire une requête SQL qui renvoie les noms et prénoms des dix acteurs qui ont tourné plus de cinq films. Renvoyer aussi le nombre de films et trier par ordre décroissant du nombre de films.

```sql
SELECT a.nom,
    a.prenom,
    count(DISTINCT r.idFilm) as nb
FROM artistes a
    JOIN roles r ON a.idArtiste = r.idActeur
GROUP BY a.idArtiste
HAVING nb >= 5
ORDER BY nb DESC
```

Écrire une requête SQL qui renvoie les pays (codePays) qui ont produit au moins $3$ drames. Renvoyer aussi le nombre de drames produits et classer selon ce nombre dans l'ordre décroissant.

```sql
SELECT DISTINCT codePays,
    count(idFilm) as nb
FROM films
WHERE genre = "Drame"
group by codePays
HAVING nb >= 3
ORDER BY nb DESC
```

Écrire une requête SQL qui renvoie les artistes qui sont à la fois acteurs et réalisateurs.

```sql
SELECT a.nom,
    a.prenom
FROM artistes a
    JOIN (
        SELECT DISTINCT idRealisateur as id
        FROM films
        INTERSECT
        SELECT DISTINCT idActeur
        FROM roles
    ) b ON a.idArtiste = b.id
ORDER BY a.nom
```

Écrire une requête SQL qui renvoie les artistes qui sont à la fois acteurs et réalisateurs mais sur un même film.

```sql
SELECT nom,
    prenom
FROM artistes
WHERE idArtiste in (
        SELECT DISTINCT f.idRealisateur
        FROM films f
            JOIN roles r ON f.idFilm = r.idFilm
        WHERE f.idRealisateur = r.idActeur
    )
```

Écrire une requête SQL qui renvoie les noms et prénoms des acteurs qui ont tourné des films dans au moins $4$ genres différents. Renvoyer aussi le nombre de genres différents et trier selon ce nombre décroissant.

```sql
SELECT a.nom,
    a.prenom,
    b.nb
FROM artistes a
    JOIN (
        SELECT r.idActeur,
            count(DISTINCT f.genre) as nb
        FROM films f
            JOIN roles r ON f.idFilm = r.idFilm
        GROUP BY r.idActeur
    ) b ON a.idArtiste = b.idActeur
WHERE b.nb >= 4
ORDER BY b.nb DESC
```
